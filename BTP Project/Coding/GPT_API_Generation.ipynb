```python
import os
from openai import OpenAI

#write your openai api key
os.environ['OPENAI_API_KEY'] = '##############'

client = OpenAI(
    # This is the default and can be omitted
    api_key=os.environ.get("OPENAI_API_KEY"),
)
text1__ = 'age variations cortical membranes'

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": " I am giving you some short text document and i want you to extend it with the same context "+ text1__,
        }
    ],
    model="gpt-3.5-turbo",
)

# Print the model's response
new_text = chat_completion.choices[0].message.content
print(new_text)
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")

```
